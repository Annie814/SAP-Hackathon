```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime
```

```python
Table_booking = pd.read_csv("Calendar.csv")

s = pd.to_datetime(Table_booking['DateStart'])
df = s.groupby(s.dt.floor('d')).size().reset_index(name='Number of booking tables')

num_Tables = df['Number of booking tables']
dates = df['DateStart']

plt.plot(dates,num_Tables)
plt.xlabel('Time')
plt.ylabel('The number of booking tables')
plt.show()
```
